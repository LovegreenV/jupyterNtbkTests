```python
from operator import itemgetter
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
f = open('/Users/vlovegreen/Desktop/BIA385/Nebraska Positive Cases Per Date.txt')
lines = f.readlines()
f.close()

# Iterate over each line and remove the new line character in the string

total = 0
lines = [s.replace('\n','') for s in lines]
for ln in lines:
    total += int(ln)

print('Total positive', total)

# Compute average
avg = total/len(lines)
print('Average value:', avg)

#make all items in the list ints
for i in range(len(lines)):
    lines[i] = int(lines[i])

#find max
maxi = max(lines)

#find index of max
j = 0
for i in lines:
    if maxi == i:
        index = j
    else:
        j = j + 1
print(maxi)
print(index)

print(lines[index])

print(lines)
```

```python
third = pd.read_csv('Desktop/BIA385/NewYork_Nebraska_PositiveCases.csv')

#show first 5 and last 5 rows
print(third.head())
print(third.tail())

NEtotal = sum(third['NEW_COVID_CASE_COUNT_Nebraska'])
NYtotal = sum(third['NEW_COVID_CASE_COUNT_NewYork'])
print('Nebraska Total:',NEtotal)
print('New York Total:',NYtotal)


numbers = (NEtotal, NYtotal)
states = ('Nebraska', 'New York')

#create the Pie chart

plt.pie(numbers, labels=states, autopct='%.1f%%')
plt.show
```

```python
#create two line charts and Ill use subplots to do it
day = range(1,37)
plt.subplot(2,1,1)
plt.plot(day,third['NEW_COVID_CASE_COUNT_Nebraska'])
plt.xlabel('Day')
plt.ylabel('Cases')
plt.title("Nebraska")
plt.subplot(2,1,2)
plt.plot(day,third['NEW_COVID_CASE_COUNT_NewYork'])
plt.xlabel('Day')
plt.ylabel('Cases')
plt.title("New York")
```
