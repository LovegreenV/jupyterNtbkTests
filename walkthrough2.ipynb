```python
nba_series = pd.read_csv('nba.csv', usecols=['Name','Salary'], index_col="Name", squeeze=True)
nba_series
```

```python
#write a function that returns average values
sumPay = nba_series.sum()
total = len(nba_series)
average = sumPay / total
def averages(n):
    if n == average:
        return "average"
    elif n > average:
        return "above average"
    elif n < average:
        return "below average"
    elif n == np.nan:
        return "No Salary Reported"
nba_series.apply(averages)
```

```python
#get a sample of 80%
all_nba = pd.read_csv("nba.csv", usecols=['Name','Team','Number','Position','Age','Height','Weight','College','Salary'], index_col="Name")
sample = all_nba.sample(frac=0.8, replace=True, random_state = 2)
sample
```

```python
#compute nan values
sample.isnull().sum(axis = 0)
#axis of 1 shows by player
```

```python
#no rows have all columns that = 0
#get mode for colleges
nbaMode = sample["College"].mode()
sample['College'] = sample['College'].replace(np.nan,"Kentucky") 
sample
```

```python
#replace salary with average
nbaAvg = sample["Salary"].mean()
sample['Salary'] = sample['Salary'].replace(np.nan,nbaAvg)
sample
```
